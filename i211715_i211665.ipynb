```python
import json
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
from cleantext import clean
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
file_path = 'movie_dataset.csv'

data = pd.read_csv(file_path)

data
```

```python
data.drop_duplicates(subset='plot_synopsis', inplace=True)
```

```python
data
```

```python
def cleaning(text):
    text=clean(text,no_urls=True,replace_with_url="")
    words = word_tokenize(text)
    alphaOnly = [word for word in words if word.isalpha()]
    Low = [word.lower() for word in alphaOnly]
    stWords = set(stopwords.words('english'))
    cleanText = [word for word in Low if word not in stWords]
    cleanText=" ".join(cleanText)
    
    return cleanText
```

```python
data['cleaned_text'] = data['plot_synopsis'].apply(lambda x: cleaning(x))
```

```python
data
```

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(lemmatized_tokens)
```

```python
data['lemmatized_text'] = data['cleaned_text'].apply(lambda x: lemmatize_text(x))
```

```python
data=pd.read_csv('moviesynopsis.csv')
```

```python
data['tags']=data['tags'].str.split(',')
```

```python
specified_genres = ['violence', 'comedy', 'cult', 'murder', 'mystery']
```

```python
def get_rows_for_genre(genre):
    genre_count = 0
    selected_rows = []
    for index, row in data.iterrows():
        if genre in row['tags'] and genre_count < 5:
            selected_rows.append(row)
            genre_count += 1
    return selected_rows

# Filtering the DataFrame based on specified genres and limit to 5 rows per genre
selected_rows = []
for genre in specified_genres:
    selected_rows.extend(get_rows_for_genre(genre))

# Creating a DataFrame with selected columns ('title', 'plot', 'genre') from the filtered rows
result_df = pd.DataFrame(selected_rows, columns=['title', 'lemmatized_text', 'tags'])
```

```python
for i in result_df.index:
    result_df['tags'].loc[i] = result_df['tags'].loc[i][0]
```

```python
result_df
```

```python
result_df.to_csv('filtered_movies.csv', index=False)
```
